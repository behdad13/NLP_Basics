## imputation

```python
data = pd.read_csv('nba.csv')
data.isnull().sum()
```

```python
x = 0 #impute it with any value
data.fillna(x)
```

```python
data.College.fillna('xyz') #impute with a value
```

```python
#impute by mean

avg_value = data.Salary.mean()
data.Salary.fillna(avg_value)
```

## Lambda Function in a dataset

```python
x = lambda a: a*a
x(3)
```

```python
#other way to impute the null value
mean_weight = data.Weight.mean()
data.Weight.apply(lambda x: x if x==x else mean_weight)

# While you work with datasets, the lambda works with apply (apply the function row by row)
```
